```python
import pickle
seed = 0
env_name = 'kangaroo'
steps_list = list(range(0, 16000001, 4000000))
weights_list = []
action_values_list = []
for steps in steps_list:
    with open(f'../stats/{env_name}/weights_steps{steps}_seed{seed}.pkl', 'rb') as f:
        weights = pickle.load(f)
        weights_list.append(weights)

    with open(f'../stats/{env_name}/action_values_steps{steps}_seed{seed}.pkl', 'rb') as f:
        action_values = pickle.load(f)
        action_values_list.append(action_values)
        
results = []
for i, steps in enumerate(steps_list):
    weights = steps_list[i]
    action_values = action_values_list[i]
    
    neural_counter = 0
    logic_counter = 0
    for v1, v2 in action_values:
        # if neural wins
        if v1 > v2:
            neural_counter += 1
        elif v2 > v1:
            logic_counter += 1
    # normalize
    neural_ratio = neural_counter / (neural_counter + logic_counter)
    logic_ratio = logic_counter / (neural_counter + logic_counter)
    results.append([neural_ratio, logic_ratio])
    
    
import pandas as pd
df_kangaroo = pd.DataFrame(results, columns=['neural', 'logic'])
```

```python
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid", {'grid.linestyle': '--'})
colors = ['hotpink', 'dodgerblue']
bluepink_cmap = LinearSegmentedColormap.from_list('bluepink', colors, N=100)
```

```python
def human_format(num):
    num = float('{:.3g}'.format(num))
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    return '{}{}'.format('{:f}'.format(num).rstrip('0').rstrip('.'), ['', 'K', 'M', 'B', 'T'][magnitude])
```

```python

fig, axes = plt.subplots(1, 2, figsize=(15, 1.8))
ax = axes[0]
ax = df_kangaroo.plot(ax=ax, kind='bar', stacked=True, colormap=bluepink_cmap, legend=False)#, figsize=(5, 2))
ax.set_xticklabels([human_format(s) for s in steps_list])
ax.set_xlabel("training steps", fontsize=16)
ax.set_ylabel("ratio", fontsize=16)
ax.set_title("Neural / Logic Ratio  (Kangaroo)", fontsize=16)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=14)
# ax.set_yticklabels(ax.get_yticklabels(), fontsize=14)

ax = axes[1]

ax = df_seaquest.plot(ax=ax, kind='bar', stacked=True, colormap=bluepink_cmap)#, figsize=(5, 2))
ax.set_xticklabels([human_format(s) for s in steps_list])
ax.set_xlabel("training steps", fontsize=16)
ax.set_ylabel("ratio", fontsize=16)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=14)
# ax.set_yticklabels(ax.get_yticklabels(), fontsize=14)
ax.set_title("Neural / Logc Ratio  (Seaquest)", fontsize=16)
ax.legend(fontsize=16, bbox_to_anchor=(1.04, 0), loc="lower left", borderaxespad=0)

path = f'figures/blender_ratio_comparison_all.pdf' 
plt.savefig(path, bbox_inches='tight')
```
