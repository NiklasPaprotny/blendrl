```python
import pandas as pd
import pickle
from termcolor import colored
import numpy as np
```

```python
pkl_report_file = "../reports/all_games_report.pkl"
latex_report_file = "../reports/all_games_report.tex"
with open(pkl_report_file, "rb") as savefile:
    df = pickle.load(savefile)
```

```python
df = df[~df.index.duplicated(keep='last')]
nb_games = len(df)
print(colored(f"{nb_games} out of 35 done !"))

# df.fillna("N/A", inplace=True)
dfT = df.T
# dfT["mean"] = dfT.mean(axis=1)
df = dfT.T

df = df.round(1)

dfT = dfT.drop(index=("C51"))
dfT = dfT.drop(index=("Random"))

# dfT["Scale"] = np.array([100, 66, 33, 0])
dfT = dfT.reset_index(level=[0]).drop(["level_0"], axis=1)
cols = dfT.columns.tolist()
capnames = {}
for col in cols:
    if "ALE" in col:
        bcol = col[4:]
    else:
        bcol = col
    if len(col) > 6:
        capnames[col] = bcol[0:5] + "."
    else:
        capnames[col] = bcol
dfT = dfT.rename(columns=capnames)
dfT = dfT.reindex(sorted(dfT.columns), axis=1)
dfT
```

```python
styler = dfT.style
styler.background_gradient(cmap='Spectral', vmin=0, vmax=100,)
styler.format(precision=1, na_rep="N/A")
styler.highlight_null("gray")
styler.map_index(
    lambda v: "rotatebox:{90}--rwrap--latex;", level=0, axis=1
)
styler.set_properties(**{'width': '3px'})
styler.format('', na_rep='MISS')  
ltx_code_o = styler.to_latex(
    caption=f"Statistiques on all games",
    clines="skip-last;data",
    convert_css=True,
    position_float="centering",
    hrules=True,
    multicol_align="c"
)
```

```python
styler
```

```python

ltx_code = ltx_code_o.replace("100.0", "100").replace("\n & 0 \\\\\n\\midrule", "")
replacements = [["ChopperCommand", "ChopperC."], ["SpaceInvaders", "SpaceInv."], ["MontezumaRevenge", "MontezumaR."], ["\\midrule", ""],
               ["\\bottomrule", ""], ["\\toprule", ""], ["Random", "\\rotatebox{90}{Random}"], ["DQN", "\\rotatebox{90}{DQN}"], 
               ["\centering", "\centering \n \setlength{\\tabcolsep}{3pt}"], ["\\cline{1-44}", "",], 
                # ["rrrrrrrrrr}", "rrrrrrrrr||c}"],
               ["precision", "Prec."], ["recall", "Rec."], ["f-score", "F1"], ["iou", "IOU"], 
               # ["{\cellcolor[HTML]{5E4FA2}} \color[HTML]{F1F1F1}  \\", "{\cellcolor[HTML]{5E4FA2}} \color[HTML]{F1F1F1} 100 \\"], 
               # ["{\cellcolor[HTML]{9E0142}} \color[HTML]{F1F1F1}  \\", "{\cellcolor[HTML]{9E0142}} \color[HTML]{F1F1F1} 0 \\"],
               ["{\cellcolor[HTML]{EEF8A4}} \color[HTML]{000000}  \\\\",  "{\cellcolor[HTML]{EEF8A4}} \color[HTML]{000000}  \\\\ \cline{1-" + str(nb_games) + "}"] 
               ]
for rpl in replacements: 
    ltx_code = ltx_code.replace(*rpl)

print(ltx_code)
```

```python
with open(latex_report_file, 'w') as texfile:
    texfile.write(ltx_code)
```
